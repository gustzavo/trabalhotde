## Exercisio B

Calcule manualmente o limite para as funções:

$$\displaystyle\lim_{x \rightarrow 0}=\frac{f(a+h)-f(a)}{h}$$

1.$f_{(x)}= 2x$

R:$f_{(x)}=\frac{f(a+h) -f(a)}{h} =\frac{2(a+h) - 2(a)}{h} = \frac{2a + 2h -2a}{h} =\frac{2h}{h}=2$

2.$f_{(x)}= x^2$

R:$f_{(x)}=\frac{f(a+h) - f(a)}{h} = \frac{(a+h)^2 - (a)^2}{h} = \frac{a^2+2ah+h^2 - a^2}{h} = \frac{2ah+h^2}{h} = \frac{h(2a+h)}{h} = 2a+h$

3.$f_{(x)}= x^2+1$

R:$f_{(x)}=\frac{f(a+h) -f(a)}{h} =\frac{(a+h)^2+1 -(a)^2+1}{h}=\frac{a^2+2ah+h^2+1-a^2-1}{h}=\frac{2ah+h^2}{h}=\frac{h(2a+h)}{h}=2a+h$

## Funcional
- Evita de usar variaveis(dados mutaveis/ estados), como por exemplo, o calculo lambda.
- Sem atribuiçoes explicitas, mas dentro por exemplo de uma função, pode haver, mas não explicita.
### Vantagens 
- O processo de alocação de memoria é automático não sendo declarado pelo usuário, eliminando os erros de alocação na função, desta maneira garantindo que a saída da função será sempre o mesmo se dado um mesmo conjunto de dados.
### Desvantagens
- Na programação funcional alguns construtos básicos sobre programação, como alocamento explícito de memória ou de variáveis, são dados como inexistentes, entretanto estes construtos ainda funcionam, porém de forma menos explícita.

```python
a=lambda f,a,h:f*(a+h)-f*(a)/h
formula=lambda x:x**3-1

def gx(formula,a,h):
    if h==0:
        return 0 
    return ((formula(a+h)-formula(a))/h)

if gx(formula,10,5)==0:
    print("não da pra dividir por 0")
else:
    print(gx(formula,10,5))
```
